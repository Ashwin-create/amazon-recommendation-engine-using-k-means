```
# Import pandas library
import pandas as pd

# Read the Excel file into a DataFrame
df = pd.read_excel("C:/Users/hp/Downloads/Cleaned_Dataset.xlsx")

# Display the DataFrame
df
```

```
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram
for col in df.select_dtypes(include=['number']).columns:
    plt.figure()
    sns.histplot(df[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.show()

# Box Plot
for col in df.select_dtypes(include=['number']).columns:
    plt.figure()
    sns.boxplot(y=df[col])
    plt.title(f'Box Plot of {col}')
    plt.show()
```

```
# Scatter Plot
for col1 in df.select_dtypes(include=['number']).columns:
    for col2 in df.select_dtypes(include=['number']).columns:
        if col1 != col2:
            plt.figure()
            sns.scatterplot(data=df, x=col1, y=col2)
            plt.title(f'Scatter Plot between {col1} and {col2}')
            plt.show()
```

```
# Box Plot or Violin Plot
top_15_asins = df['asin'].value_counts().head(15).index

# Filtering the dataframe
filtered_df = df[df['asin'].isin(top_15_asins)]

# Create Box Plot or Violin Plot
for num_col in filtered_df.select_dtypes(include=['number']).columns:
    plt.figure(figsize=(12, 6))
    sns.boxplot(x=filtered_df['asin'], y=filtered_df[num_col])
    plt.xticks(rotation=45)
    plt.title(f'Box Plot of {num_col} by asin (Top 15)')
    plt.show()
```

```
# Pair Plot
sns.pairplot(df.select_dtypes(include=['number']))
plt.show()
```

```
# Heatmap of Correlation Matrix
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```
# Univariate Analysis - Categorical Data (Bar Charts)
categorical_columns = ['brand', 'category']

for col in categorical_columns:
    plt.figure(figsize=(10, 6))
    top_categories = df[col].value_counts().head(10)  # top 10 categories
    sns.barplot(x=top_categories.values, y=top_categories.index)
    plt.title(f'Bar Chart of Top 10 {col.capitalize()}')
    plt.show()

```

```
# Selecting top 10 brands for clarity
top_brands = df['brand'].value_counts().head(10).index

# Filtering data for only top brands
filtered_data = df[df['brand'].isin(top_brands)]

# Crosstabulation
crosstab = pd.crosstab(filtered_data['brand'], filtered_data['category'])

# Stacked Bar Chart
crosstab.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Stacked Bar Chart of Categories within Top 10 Brands')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```
import seaborn as sns

# Selecting numerical columns for pair plot and heatmap
numerical_columns = df.select_dtypes(include=['number']).columns

# Pair Plot
sns.pairplot(df[numerical_columns])
plt.show()

# Heatmap of Correlation Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(df[numerical_columns].corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Time Series Plot
# Convert 'reviewTime' to datetime if necessary
df['reviewTime'] = pd.to_datetime(df['reviewTime'], format='%Y') # Assuming 'reviewTime' is a year

plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='reviewTime', y='overall_rating')
plt.title('Time Series Plot of Overall Rating Over Time')
plt.show()

# Heatmap for Categorical Data (using top categories for clarity)
top_brands = df['brand'].value_counts().head(10).index
top_categories = df['category'].value_counts().head(10).index

filtered_data = df[df['brand'].isin(top_brands) & df['category'].isin(top_categories)]
ctab = pd.crosstab(filtered_data['brand'], filtered_data['category'])
plt.figure(figsize=(12, 8))
sns.heatmap(ctab, annot=True, cmap='coolwarm')
plt.title('Heatmap of Top Brands and Categories')
plt.show()

# Faceted Plot
# Here we use FacetGrid to create a series of plots for 'overall_rating' across different 'brands'
g = sns.FacetGrid(df[df['brand'].isin(top_brands)], col='brand', col_wrap=3, height=4)
g.map(sns.histplot, 'overall_rating')
plt.show()
```
